```python
def within_sum_of_squares(data, centroids, labels):
    SSW = 0
    for l in np.unique(labels):
        data_l = data[labels == l]
        resid = data_l - centroids[l]
        SSW += (resid**2).sum()
    return SSW
```

```python
from sklearn.cluster import KMeans
wss_list = []
for i in range(2, 10+1):
  print('Training {} cluster algo...'.format(i))
  km = KMeans(n_clusters=i)
  km.fit(location)
  wss = within_sum_of_squares(np.array(location),km.cluster_centers_, km.predict(location))    
  wss_list.append(wss)
plt.plot(wss_list)
plt.title('WSS Plot')
plt.xlabel('# of Clusters')
plt.ylabel('WSS')
plt.show()
```

```python
from scipy.spatial.distance import cdist

#find the minimumn distance in points list
def points(distances):
  return np.array([np.argmin(d) for d in distances])

#stacking in row-wise
def stacking(centriods):
  return np.vstack(centriods)

def kmeans(data,k, iterations=100):
  random_index = np.random.choice(len(data), k, replace=False)
  #Randomly choosing Centroids 
  cent = data[random_index, :]
    
  #distance beteween each data and each centriods
  dis = cdist(data, cent,'euclidean')
  
  #Centroid with the minimum Distance
  pointsList = points(dis) 
  
  for it in range(iterations): 
    cent = []
    #put centriods to where it belongs to
    for i in range(k):
    
      newly_cent = data[pointsList== i].mean(axis=0) 
      cent.append(newly_cent)

    cent = stacking(cent)
    dist = cdist(data, cent, 'euclidean')
    
    pointsList = points(dist)
        
  return pointsList
```

```python
from sklearn.decomposition import PCA
pca2 = PCA(2)
n_clusters = 6
#Transform the data
data2 = pca2.fit_transform(df)
 
#Applying our function
label = kmeans(data2,n_clusters)
 
#Visualize the results
plt.figure(figsize=(10,10))
u_labels2 = np.unique(label)
for i in u_labels2:
    plt.scatter(data2[label == i , 0] , data2[label == i , 1] , label = i)

# plt.legend()
plt.title("Clusters of OK Cupid online dating app profile Recommender", fontsize=30)
plt.xlabel('PCA 1', fontsize=20)
plt.ylabel('PCA 2', fontsize=20)

plt.gca().legend(('Cluster 0','Cluster 1', 'Cluster 2','Cluster 3','Cluster 4','Cluster 5'),prop={"size":20})
plt.show()
```
